## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

Electricity demand depends on many interacting factors such as weather conditions, seasonal patterns, holidays, and historical consumption. These relationships are nonlinear and constantly changing.

A fixed rule like:

"If temperature is high, demand increases"

cannot capture all these interactions.

Machine learning learns these complex patterns from historical data and adapts to new situations, leading to more accurate predictions than manually written rules.
